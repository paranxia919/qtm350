# Group Members Assignment

This notebook contains the existing student groups and will randomly assign the remaining students to new groups.

```python
# Import required libraries
import random
import pandas as pd
from IPython.display import Markdown, display
```

```python
# Define existing groups
existing_groups = {
    "Group 01": [
        "Laura Zang",
        "Tiffany Wickens",
        "Chloe Guerrero",
        "Mark Arshavsky",
        "Chloe Zhao"
    ],
    "Group 02": [
        "Vicki Wang",
        "Gloria Wang",
        "Annie Cao",
        "Yan Yang"
    ],
    "Group 03": [
        "Anita Osuri",
        "Laura Wang",
        "Winnie Lau",
        "Caitlyn Ye",
        "Tashfia Noor" 
    ],
    "Group 04": [
        "Leo Xu",
        "Lucy Liu",
        "Tom Suo",
        "Mike Wu",
        "Santiago Vazquez"
    ],
    "Group 05": [
        "Francisco Lyu",
        "Jinhao Tian",
        "Jonathan Wang",
        "Tiffany Yang"
    ]
}

# Complete student list from the previous analysis
all_students = [
    "Mark Arshavsky",
    "Sohan Bellam",
    "Krisha Bhai",
    "Annie Cao",
    "Yuting Chen",
    "Christopher Chong",
    "Chloe Guerrero",
    "Ari Gurovich",
    "Justin He",
    "Maanya Ladhani",
    "Winnie Lau",
    "Bert Li",
    "Sherelle Li",
    "Chang Liu",
    "Emily Liu",
    "Lucy Liu",
    "Francisco Lyu",
    "Pak Huen Myron Ngan",
    "Tashfia Noor",
    "Anita Osuri",
    "Krishang Shah",
    "Sally Shen",
    "Sriram Subramoniapillai",
    "Tom Suo",
    "Jinhao Tian",
    "Santiago Vazquez",
    "Gloria Wang",
    "Jonathan Wang",
    "Laura Wang",
    "Vicki Wang",
    "Tiffany Wickens",
    "Mike Wu",
    "Leo Xu",
    "Tiffany Yang",
    "Tristan Yang",
    "William Yang",
    "Yan Yang",
    "Caitlyn Ye",
    "Laura Zang",
    "Sherry Zhang",
    "Chloe Zhao"
]
```

```python
# Find students who haven't been assigned to groups yet
assigned_students = []
for group_members in existing_groups.values():
    assigned_students.extend(group_members)

unassigned_students = [student for student in all_students if student not in assigned_students]

print(f"Total students: {len(all_students)}")
print(f"Assigned students: {len(assigned_students)}")
print(f"Unassigned students: {len(unassigned_students)}")
print("\nUnassigned students:")
for student in unassigned_students:
    print(f"- {student}")
```

```python
# Randomly assign remaining students to groups
random.shuffle(unassigned_students)

# Create new groups (starting from Group 06)
new_groups = {}
group_number = 6

# Create groups of 4-5 students, ensuring no group has more than 5
i = 0
while i < len(unassigned_students):
    remaining_students = len(unassigned_students) - i
    
    # Determine optimal group size
    if remaining_students >= 9:  # If we have 9 or more left, make groups of 5
        group_size = 5
    elif remaining_students >= 5:  # If we have 5-8 left, make groups of 4
        group_size = 4
    elif remaining_students == 8:  # Special case: split 8 into two groups of 4
        group_size = 4
    else:  # 1-4 students left, put them all in one group
        group_size = remaining_students
    
    group_key = f"Group {group_number:02d}"
    group_members = unassigned_students[i:i+group_size]
    new_groups[group_key] = group_members
    
    print(f"Created {group_key} with {len(group_members)} members")
    
    i += group_size
    group_number += 1

# Combine existing and new groups
all_groups = {**existing_groups, **new_groups}

print("\nGroup assignment completed!")
for group_name, members in all_groups.items():
    print(f"\n{group_name} ({len(members)} members):")
    for member in members:
        print(f"  - {member}")
```

```python
# Create the final markdown table
def create_email(name):
    # Convert name to email format (firstname.lastname@emory.edu)
    parts = name.lower().split()
    if len(parts) >= 2:
        # Handle special cases
        if "(" in parts[1]:
            return f"{parts[0].lower()}.{parts[1].split('(')[0].lower()}@emory.edu"
        # Handle compound last names
        elif len(parts) > 2 and parts[-1] in ['pillai']:
            return f"{parts[0].lower()}.{parts[-2].lower()}.{parts[-1].lower()}@emory.edu"
        elif parts[-1] == "ngan":
            return f"{'.'.join(parts[:-1])}.{parts[-1]}@emory.edu"
        else:
            return f"{parts[0].lower()}.{parts[-1].lower()}@emory.edu"
    return name.lower() + "@emory.edu"

def get_surname(name):
    """Extract surname for sorting"""
    parts = name.split()
    if len(parts) >= 2:
        # Handle special cases like "Pak Huen Myron Ngan"
        if len(parts) > 2 and parts[-1] in ['Ngan', 'Subramoniapillai']:
            return parts[-1]
        # Handle names with parentheses
        if "(" in parts[-1]:
            return parts[-2]
        return parts[-1]
    return name

# Create markdown table
table_lines = []
table_lines.append("| Name | SIS ID | Group Number | Email |")
table_lines.append("|------|--------|--------------|------|")

# SIS IDs mapping (from the original data)
sis_ids = {
    "Mark Arshavsky": "2556041",
    "Sohan Bellam": "2546700",
    "Krisha Bhai": "2579706",
    "Annie Cao": "2599315",
    "Yuting Chen": "2554090",
    "Christopher Chong": "2582000",
    "Chloe Guerrero": "2578787",
    "Ari Gurovich": "2586399",
    "Justin He": "2549590",
    "Maanya Ladhani": "2549003",
    "Winnie Lau": "P7319615",
    "Bert Li": "2549011",
    "Sherelle Li": "2583169",
    "Chang Liu": "2579008",
    "Emily Liu": "2550121",
    "Lucy Liu": "2561533",
    "Francisco Lyu": "2549030",
    "Pak Huen Myron Ngan": "2559660",
    "Tashfia Noor": "2588861",
    "Anita Osuri": "2557540",
    "Krishang Shah": "2579276",
    "Sally Shen": "2549393",
    "Sriram Subramoniapillai": "2591970",
    "Tom Suo": "2549787",
    "Jinhao Tian": "2562165",
    "Santiago Vazquez": "2592291",
    "Gloria Wang": "2550999",
    "Jonathan Wang": "2549119",
    "Laura Wang": "2550326",
    "Vicki Wang": "2581225",
    "Tiffany Wickens": "2537894",
    "Mike Wu": "2549432",
    "Leo Xu": "2592825",
    "Tiffany Yang": "2551053",
    "Tristan Yang": "2592886",
    "William Yang": "2487074",
    "Yan Yang": "2610419",
    "Caitlyn Ye": "2555901",
    "Laura Zang": "2549139",
    "Sherry Zhang": "2593324",
    "Chloe Zhao": "2593352"
}

# Collect all students with their group numbers
all_students_with_groups = []
for group_name, members in all_groups.items():
    group_num = group_name.split()[1]  # Extract group number
    for student in members:
        sis_id = sis_ids.get(student, "")
        email = create_email(student)
        surname = get_surname(student)
        all_students_with_groups.append((surname, student, sis_id, group_num, email))

# Sort by surname alphabetically
all_students_with_groups.sort(key=lambda x: x[0])

# Add sorted students to the table
for surname, student, sis_id, group_num, email in all_students_with_groups:
    table_lines.append(f"| {student} | {sis_id} | {group_num} | {email} |")

# Display the markdown table
markdown_table = "\n".join(table_lines)
display(Markdown(markdown_table))
```
